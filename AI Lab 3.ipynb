```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=var[['weight']].values #input/independent var[[]] nibo cz input can be multiple
y=var['price']           #output/dependent var follows np
```

```python
x[0:10]
```

```python
x
```

```python
x.max()
```

```python
plt.scatter(x,y,marker='+',color='blue')
plt.xlabel("Weight")
plt.ylabel("Price")
plt.title("Price prediction")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=1)
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression #(onno kono algo use krlemodel name cng hoto)
```

```python
reg=LinearRegression() #create obj & fit LinearRegression() algo in it
```

```python
reg.fit(x_train, y_train) #train through LinearRegression() algo
```

```python
reg.predict(x_test) #test, input train krbo ,mechine output predict krbo
```

```python
y_test #actual value of output...%pite obj.score()
```

```python
plt.plot(var['weight'],reg.predict(var[['weight']]))
```

```python
plt.scatter(x,y, marker='+',color='red')
plt.xlabel("Weight")
plt.ylabel("Price")
plt.title("Price prediction")
plt.plot(var['weight'],reg.predict(var[['weight']]))
#pic of diff from real val(+ dia predicted val)...curve using y=mx+c[y=price,x=weight]
```

```python
reg.intercept_ #y=mx+c er c holo intercept_
```

```python
reg.coef_  #y=mx+c er m holo coefficient
```

```python
reg.predict(x_test)
```

```python
y_test
```

```python
reg.score(x_test,y_test) #diff between main & predicted accuracy
```
